```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path("..").resolve()
sys.path.insert(0, str(ROOT))

from custom.cotsfa_original import (
    CoTSFAAugmentConfig,
    cotsfa_anomaly_curve,
    apply_cotsfa_curve_once,
    generate_cotsfa_augmentations,
)

print("ROOT:", ROOT)
```

```python
csv_path = ROOT / "dataset" / "ETT-small" / "ETTh1.csv"
df = pd.read_csv(csv_path)

cols = ["HUFL", "HULL", "MUFL", "MULL", "LUFL", "LULL", "OT"]

seq_len = 96
pred_len = 96
start = 0

values = df[cols].values.astype(np.float32)

x = values[start : start + seq_len]
y = values[start + seq_len : start + seq_len + pred_len]

print("x shape:", x.shape)
print("y shape:", y.shape)

print("ETTh1 x min:", x.min())
print("ETTh1 x max:", x.max())
print("ETTh1 x mean:", x.mean())
print("ETTh1 x std:", x.std())
```

```python
config_raw = CoTSFAAugmentConfig(
    num_augmentations=5,
    amplitude_mean=74120.0,
    amplitude_std=20000.0,
    exponent_mean=0.806,
    exponent_std=0.2,
    normalize_curve=False,
    scaling_mode="raw",
    strength=1.0,
)

curve = cotsfa_anomaly_curve(
    length=96,
    amplitude=74120.0,
    exponent=0.806,
    b=config_raw.b,
    z=config_raw.z,
    normalize=config_raw.normalize_curve,
)

print("curve min:", curve.min())
print("curve max:", curve.max())
print("curve t=30 value:", curve[29])
print("valid max < 2.0:", curve.max() < 2.0)
print("valid t30 < 0.4:", curve[29] < 0.4)

plt.figure(figsize=(10, 3))
plt.plot(curve)
plt.axhline(2.0, linestyle="--", label="max constraint 2.0")
plt.axhline(0.4, linestyle="--", label="t=30 constraint 0.4")
plt.axvline(29, linestyle="--", label="t=30")
plt.title("Original Co-TSFA raw anomaly curve")
plt.xlabel("t")
plt.ylabel("a(t)")
plt.legend()
plt.show()
```

```python
x_io_raw, y_io_raw, meta_io_raw = apply_cotsfa_curve_once(
    x=x,
    y=y,
    mode="input_only",
    config=config_raw,
    seed=42,
)

print(meta_io_raw)
print("x changed:", np.mean(np.abs(x_io_raw - x)))
print("y changed:", np.mean(np.abs(y_io_raw - y)))
```

```python
x_oo_raw, y_oo_raw, meta_oo_raw = apply_cotsfa_curve_once(
    x=x,
    y=y,
    mode="input_output",
    config=config_raw,
    seed=43,
)

print(meta_oo_raw)
print("x changed:", np.mean(np.abs(x_oo_raw - x)))
print("y changed:", np.mean(np.abs(y_oo_raw - y)))
```

```python
channel_idx = cols.index("OT")

fig, axes = plt.subplots(2, 1, figsize=(12, 7), sharex=False)

axes[0].plot(x[:, channel_idx], label="x original")
axes[0].plot(x_io_raw[:, channel_idx], label="x input-only raw augmented")
axes[0].set_title("Input-only raw Co-TSFA curve on OT")
axes[0].legend()

combined_original = np.concatenate([x[:, channel_idx], y[:, channel_idx]])
combined_aug_raw = np.concatenate([x_oo_raw[:, channel_idx], y_oo_raw[:, channel_idx]])

axes[1].plot(combined_original, label="original x+y")
axes[1].plot(combined_aug_raw, label="input-output raw augmented x+y")
axes[1].axvline(seq_len, linestyle="--", label="forecast boundary")
axes[1].set_title("Input-output raw Co-TSFA curve on OT")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
config_std = CoTSFAAugmentConfig(
    num_augmentations=5,
    amplitude_mean=74120.0,
    amplitude_std=20000.0,
    exponent_mean=0.806,
    exponent_std=0.2,
    normalize_curve=False,
    scaling_mode="std",
    strength=1.0,
)

x_oo_std, y_oo_std, meta_oo_std = apply_cotsfa_curve_once(
    x=x,
    y=y,
    mode="input_output",
    config=config_std,
    seed=43,
)

print(meta_oo_std)
print("x changed std:", np.mean(np.abs(x_oo_std - x)))
print("y changed std:", np.mean(np.abs(y_oo_std - y)))
```

```python
combined_aug_std = np.concatenate([x_oo_std[:, channel_idx], y_oo_std[:, channel_idx]])

plt.figure(figsize=(12, 4))
plt.plot(combined_original, label="original x+y")
plt.plot(combined_aug_raw, label="raw augmented")
plt.plot(combined_aug_std, label="std-scaled augmented")
plt.axvline(seq_len, linestyle="--", label="forecast boundary")
plt.title("Raw vs std-scaled Co-TSFA input-output augmentation on OT")
plt.legend()
plt.show()
```

```python
x_augs, y_augs, metas = generate_cotsfa_augmentations(
    x=x,
    y=y,
    mode="input_output",
    config=config_raw,
    seed=123,
)

print("x_augs shape:", x_augs.shape)
print("y_augs shape:", y_augs.shape)
print("num metadata:", len(metas))

for m in metas:
    print({
        "idx": m["augmentation_index"],
        "start_idx": m["start_idx"],
        "curve_len": m["curve_len"],
        "curve_valid": m["curve_valid"],
        "raw_curve_max": m["raw_curve_max"],
        "raw_curve_t30": m["raw_curve_t30"],
        "scaling_mode": m["scaling_mode"],
    })
```

```python
strengths = [1.0, 3.0, 5.0]

raw_results = {}

for strength in strengths:
    config = CoTSFAAugmentConfig(
        num_augmentations=5,
        amplitude_mean=74120.0,
        amplitude_std=20000.0,
        exponent_mean=0.806,
        exponent_std=0.2,
        normalize_curve=False,
        scaling_mode="raw",
        strength=strength,
    )

    x_aug, y_aug, meta = apply_cotsfa_curve_once(
        x=x,
        y=y,
        mode="input_output",
        config=config,
        seed=43,
    )

    raw_results[strength] = {
        "x_aug": x_aug,
        "y_aug": y_aug,
        "meta": meta,
        "x_changed": np.mean(np.abs(x_aug - x)),
        "y_changed": np.mean(np.abs(y_aug - y)),
    }

    print(f"\nstrength = {strength}")
    print(meta)
    print("x changed:", raw_results[strength]["x_changed"])
    print("y changed:", raw_results[strength]["y_changed"])
```

```python
channel_idx = cols.index("OT")
combined_original = np.concatenate([x[:, channel_idx], y[:, channel_idx]])

plt.figure(figsize=(12, 4))
plt.plot(combined_original, label="original x+y")

for strength in strengths:
    x_aug = raw_results[strength]["x_aug"]
    y_aug = raw_results[strength]["y_aug"]
    combined_aug = np.concatenate([x_aug[:, channel_idx], y_aug[:, channel_idx]])
    plt.plot(combined_aug, label=f"raw strength={strength}")

plt.axvline(seq_len, linestyle="--", label="forecast boundary")
plt.title("Raw Co-TSFA input-output augmentation by strength on OT")
plt.legend()
plt.show()
```

```python
ot_std = np.std(np.concatenate([x[:, channel_idx], y[:, channel_idx]]))
ot_range = np.max(combined_original) - np.min(combined_original)

print("OT std:", ot_std)
print("OT range:", ot_range)

for strength in strengths:
    x_aug = raw_results[strength]["x_aug"]
    y_aug = raw_results[strength]["y_aug"]
    combined_aug = np.concatenate([x_aug[:, channel_idx], y_aug[:, channel_idx]])

    diff = np.abs(combined_aug - combined_original)

    print(f"\nstrength = {strength}")
    print("mean abs diff:", diff.mean())
    print("max abs diff:", diff.max())
    print("mean diff / OT std:", diff.mean() / ot_std)
    print("max diff / OT std:", diff.max() / ot_std)
    print("max diff / OT range:", diff.max() / ot_range)
```
