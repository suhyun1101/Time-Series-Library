```python
import sys
sys.path.append("..")
sys.path.append(".")

import torch

from custom.losses import CoTSFAAlignmentLoss
```

```python
B = 4
T = 96
C = 7
D = 16

z = torch.randn(B, D)
z_aug = z + 0.01 * torch.randn(B, D)

y = torch.randn(B, T, C)
y_aug_input_only = y.clone()

loss_fn = CoTSFAAlignmentLoss()

loss, sim_latent, sim_output = loss_fn(
    z=z,
    z_aug=z_aug,
    y=y,
    y_aug=y_aug_input_only,
)

print("input-only case")
print("loss:", loss.item())
print("sim_latent:", sim_latent.item())
print("sim_output:", sim_output.item())
```

```python
y_aug_changed = y + 0.5 * torch.randn(B, T, C)

loss, sim_latent, sim_output = loss_fn(
    z=z,
    z_aug=z_aug,
    y=y,
    y_aug=y_aug_changed,
)

print("input-output changed case")
print("loss:", loss.item())
print("sim_latent:", sim_latent.item())
print("sim_output:", sim_output.item())
```
